```python
import pandas as pd
```

```python
#read and display data
train = pd.read_csv("/content/drive/MyDrive/train.csv")
train
```

```python
#loại bỏ cột id do nó không ảnh hưởn đến response
train = train.drop(columns = ["id"])
train
```

```python
#vì kênh bán hàng không ảnh hưởng đến response nên ta cũng bỏ
train = train.drop(columns = ["Policy_Sales_Channel"])
```

```python
train
```

```python
#chuyển Gender thành dạng số
train = pd.get_dummies(train, columns=["Gender"])
train
```

```python
#vì cột Male và Female bù trừ cho nhau nên ta loại bỏ một cột
train = train.drop(columns = "Gender_Female")
train
```

```python
#ta số hóa cột Vehicle_Age
train = pd.get_dummies(train, columns=["Vehicle_Age"], drop_first = True)
```

```python
train
```

```python
#Vì khoảng số giữa 3 cột Age, Annual_Premium, Vintage là rất khác nhau nên ta chuyển chúng về khoảng 0-1
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
train[["Age", "Annual_Premium", "Vintage"]] = scaler.fit_transform(train[["Age", "Annual_Premium", "Vintage"]])
```

```python
train
```

```python
#ta khảo sát dữ liệu cột Region_Code
```

```python
train["Region_Code"].value_counts()
```

Như ở đây, ta thấy mã vùng thứ 28 là nhiều nhất, mà số liệu ở cột này rất lộn xộn, vì thế mình sẽ số hóa cột này bằng cách, ô nào có giá trị 28 thì để là 1, còn lại là 0

```python
train["Region_Code"] = train["Region_Code"].apply(lambda x: 1 if x==28 else 0)
```

```python
train
```

```python
train = pd.get_dummies(train, columns = ["Vehicle_Damage"], drop_first = True)
```

```python
train
```

cột input bao gồm bảng và loại bỏ cột response, còn output là response, chúng ta sẽ khảo sát sự tương quan giữa các số liệu

```python
train.corr()
```

```python
#Đồ thị hóa 
import seaborn as sb
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize = (10,10))
dataplot = sb.heatmap(train.corr(), annot = True, ax = ax)
```

```python
#tách dữ liệu x = input, y = output
x = train.drop(columns = ["Response"])
y = train["Response"]
x.shape, y.shape
```

```python
#tạo dữ liệu train, test
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
y_train.value_counts()
```

Như ở đây, kết quả đầu ra bị lệch hẳn về phía 0, nên để cho khách quan, ta sẽ dùng Oversampling sử dụng SMOTE

```python
from imblearn.over_sampling import SMOTE
sm = SMOTE()

x_train_os, y_train_os = sm.fit_resample(x_train, y_train)
x_train_os.shape, y_train_os.shape
```

như ở đây dữ liệu đã tăng lên 

```python
y_train_os.value_counts()
```

dữ liệu đã cân bằng nhau, sau đây chúng ta sẽ thử train dữ liệu với các thuật toán

Logistic without **oversampling**

```python
from sklearn.metrics import classification_report
from sklearn.metrics import plot_confusion_matrix
from sklearn.linear_model import LogisticRegression

model_log = LogisticRegression()
model_log.fit(x_train, y_train)

y_pred = model_log.predict(x_test)

print(classification_report(y_test, y_pred))
plot_confusion_matrix(model_log, x_test, y_test)
```

như ở đây, recall của kết quả 1 quá thấp khi không có OverSampling, gần như tất cả những kết quả = 1 đều bị dự đoán = 0


Logistic with **Oversampling**

```python
model_log = LogisticRegression()
model_log.fit(x_train_os, y_train_os)

y_pred = model_log.predict(x_test)

print(classification_report(y_test, y_pred))
plot_confusion_matrix(model_log, x_test, y_test)
```

Ồ, có vẻ recall đã khá ổn, ở mức 0.97 => ta có thể hiểu là 97% khách hàng có nhu cầu mua chéo đa được nhận định đúng 

Tiếp theo ta sẽ thử với thuật toán XGBOOST (no OS/ OS)

```python
#XGBOOST without oversampling
#XGBOOST sẽ gặp vấn đề với dữ liệu có tên cột chứa < > => đổi tên cột
x_train = x_train.rename(columns={"Vehicle_Age_< 1 Year" : "Vehicle_lesst_1 Year", "Vehicle_Age_> 2 Years" : "Vehicle_more_2 Year"}, errors = "raise")
x_test = x_test.rename(columns={"Vehicle_Age_< 1 Year" : "Vehicle_lesst_1 Year", "Vehicle_Age_> 2 Years" : "Vehicle_more_2 Year"}, errors = "raise")
x_train_os = x_train_os.rename(columns={"Vehicle_Age_< 1 Year" : "Vehicle_lesst_1 Year", "Vehicle_Age_> 2 Years" : "Vehicle_more_2 Year"}, errors = "raise")
```

```python
import xgboost as xgb
model_xgb = xgb.XGBClassifier(random_state = 42, n_estimators = 100)
model_xgb.fit(x_train, y_train)

y.pred = model_xgb.predict(x_test)
print(classification_report(y_test, y_pred))
plot_confusion_matrix(model_xgb, x_test, y_test)
```

có vẻ rất tồi

```python
#XGBoost with Oversampling
import xgboost as xgb
model_xgb = xgb.XGBClassifier(random_state = 42, n_estimators = 100)
model_xgb.fit(x_train_os, y_train_os)

y.pred = model_xgb.predict(x_test)
print(classification_report(y_test, y_pred))
plot_confusion_matrix(model_xgb, x_test, y_test)
```

có vẻ ổn hơn rồi
